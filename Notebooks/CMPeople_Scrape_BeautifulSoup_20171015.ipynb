```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

import csv
```

```python
pd.set_option('max_colwidth', 800)
pd.set_option('max_columns', 1000)
pd.set_option('max_rows', 1000)
```

```python
cityids = list(range(148, 349))
```

```python
# TODO: this is missing city IDs!
l = []
for x in cityids:
    url = 'http://creativemornings.com/people?sort_key=newest&city_id=' + str(x)
    response = requests.get(url)
    content = response.content
    soup = BeautifulSoup(content, "lxml")
    c = soup.find_all(attrs={'class': 'listing-count'})
    c = str(c)
    c = c.replace('[<span class="listing-count">', '')
    c = c[:-8]
    c = int((c.replace(',', '')))
    
    n = soup.find_all(attrs={'class': 'person-name'})
    
    s = soup.find_all(attrs={'class': 'person-desc'})
    
    p = soup.find_all(attrs={'class': 'avatar-bg cm_bg '})
    
    l.append({'Count': c, 'Name': n, 'Description':
        s, 'Picture': p})
    
```

```python
df = pd.DataFrame.from_records(l)
```

```python
df['Description'] = df['Description'].map(lambda x: str(x).replace('<div class="person-desc">', ''))
df['Description'] = df['Description'].map(lambda x: str(x).replace('</div>', ''))
```

```python
df['Name'] = df['Name'].map(lambda x: str(x).replace('<h1 class="person-name">', ''))
df['Name'] = df['Name'].map(lambda x: str(x).replace('<h1 class="person-name">', ''))
df['Name'] = df['Name'].map(lambda x: str(x).replace('</h1>', ''))
```

```python
# Add Picture, will need regex.
```

```python
dcount = []
for x in df['Description']:
    string = str(x)
    dlist = string.split(',')
    while ' ' in dlist:
        dlist.remove(' ')
    dc = len(dlist)
    dcount.append(dc)
    
df['dcount'] = dcount
```

```python
ncount = []
for x in df['Name']:
    string = str(x)
    dlist = string.split(',')
    while ' ' in dlist:
        dlist.remove(' ')
    nc = len(dlist)
    ncount.append(nc)
    
df['ncount'] = ncount
```

```python
pcount = []
for x in df['Picture']:
    string = str(x)
    plist = string.split(',')
    pc = len([x for x in plist if "user" in x ])
    pcount.append(pc)
    
df['pcount'] = pcount
```

```python
countdf = df[['Count', 'ncount', 'dcount', 'pcount']]
```
